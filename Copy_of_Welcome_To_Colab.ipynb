# Welcome to Colab!

```
class User:
    def __init__(self, nama, email):
        self.nama = nama  # Atribut dasar
        self.email = email # Atribut dasar

    def salam(self):
        return f"Halo, saya {self.nama} dari Panenusa."

class Admin(User):
    def __init__(self, nama, email):
        super().__init__(nama, email) # Memanggil atribut dari induk
        self.role = "Admin"

    def kelola_sistem(self):
        return "Akses diterima: Mengelola validasi data peta pangan."

class Petani(User):
    def __init__(self, nama, email):
        super().__init__(nama, email) # Memanggil atribut dari induk
        self.role = "Petani"

    def input_panen(self):
        return "Berhasil: Data hasil panen padi telah diinput."

admin_panenusa = Admin("Alisha Admin", "admin@panenusa.com")
petani_madiun = Petani("Budi Petani", "budi@gmail.com")
print("=== OUTPUT SISTEM PANENUSA ===")

print(admin_panenusa.salam()) # Memanggil fungsi warisan dari induk
print(admin_panenusa.kelola_sistem()) # Memanggil fungsi khusus admin

print("-" * 30)

print(petani_madiun.salam()) # Memanggil fungsi warisan dari induk
print(petani_madiun.input_panen()) # Memanggil fungsi khusus petani
```

Polimorfisme

```
class Transaksi:
    def hitung_layanan(self, nominal):
        pass

class AkunPetani(Transaksi):
    def hitung_layanan(self, nominal):
        return nominal * 0.02  # Potongan biaya admin 2% untuk petani

class AkunPembeli(Transaksi):
    def hitung_layanan(self, nominal):
        return 5000  # Biaya layanan tetap Rp 5.000 untuk pembeli

def proses_biaya(objek_transaksi, nominal):
    print(f"Biaya Layanan: Rp {objek_transaksi.hitung_layanan(nominal)}")

petani = AkunPetani()
pembeli = AkunPembeli()

proses_biaya(petani, 1000000)  # Hasil berbeda meskipun fungsi sama
proses_biaya(pembeli, 1000000)
```

Enkapsulasi

```
class AkunKeuangan:
    def __init__(self, nama, saldo, pin):
        self.nama = nama
        self._saldo = saldo       # Protected (disarankan tidak diakses langsung)
        self.__pin = pin           # Private (Rahasia/Encapsulated total)

    # Getter untuk akses saldo secara resmi
    def cek_saldo(self):
        return f"Saldo Anda: Rp {self._saldo}"

    # Metode untuk validasi tanpa membocorkan PIN
    def verifikasi_pin(self, input_user):
        if input_user == self.__pin:
            return "Verifikasi Berhasil!"
        else:
            return "PIN Salah!"

akun_alisha = AkunKeuangan("Alisha", 5000000, "1234")

print(f"Pemilik Akun: {akun_alisha.nama}")
print(akun_alisha.cek_saldo())

print("-" * 20)
input_pin = "1234" # Anggap saja ini input dari user
print(f"Mencoba PIN: {input_pin}")
print(akun_alisha.verifikasi_pin(input_pin))
```
