Inheritance (USER)

```
# Induk: Kategori informasi umum untuk petani
class InfoTani:
    def __init__(self, judul, tanggal):
        self.judul = judul
        self.tanggal = tanggal

# Anak 1: Informasi spesifik harga pasar
class HargaPasar(InfoTani):
    def detail(self, harga):
        return f"Update {self.judul}: Rp {harga}/kg pada {self.tanggal}"

# Anak 2: Informasi spesifik cuaca
class CuacaBMKG(InfoTani):
    def detail(self, kondisi):
        return f"Prediksi {self.judul}: {kondisi} untuk tanggal {self.tanggal}"

hargagabah = HargaPasar("Harga Gabah", "1 Mei 2026")
print(hargagabah.detail(6500))
```

Inheritance (ADMIN)

```
class KelolaAkun:
    def __init__(self, id_user):
        self.id_user = id_user

class ResetAkses(KelolaAkun):
    def eksekusi(self):
        return f"[Admin] Password untuk User {self.id_user} telah di-reset ke default."

class VerifikasiAkun(KelolaAkun):
    def eksekusi(self):
        return f"[Admin] Akun User {self.id_user} kini telah berstatus TERVERIFIKASI."

task_admin = VerifikasiAkun("USR-99")
print(task_admin.eksekusi())
```

Polimorfisme (ADMIN)

```
class LaporanTransaksi:
    def cetak(self):
        return "[Admin] Mencetak Laporan Keuangan: Total transaksi bulan ini Rp 500jt."

class LaporanLogSistem:
    def cetak(self):
        return "[Admin] Mencetak Log Aktivitas: 1.200 user berhasil login hari ini."

def klik_tombol_cetak(laporan):
    print(laporan.cetak())

klik_tombol_cetak(LaporanTransaksi())
klik_tombol_cetak(LaporanLogSistem())
```

Enkapsulasi (ADMIN)

```
class KeamananSistem:
    def __init__(self, api_key):
        self.__api_key = api_key # Private: Kunci rahasia sistem

    def cek_status_api(self, input_key):
        if input_key == self.__api_key:
            return "[Admin] Koneksi API: Terhubung Aman."
        return "[Admin] Akses API Ditolak! Kunci Salah."

sys_secure = KeamananSistem("SECRET123")
print(sys_secure.cek_status_api("SECRET123"))
```
